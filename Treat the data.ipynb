```python
import pandas as pd
import re
```

```python
#Import data
data = pd.read_csv("C:\\Users\\beatr\\Documents\\Beatriz\\Ironhack\\Projects\\Tableau-Mini-Project\pizza_v1.csv")
```

```python
data.head(10)
```

```python
#Update the column of the price using REGEX
data['price_rupiah'] = data['price_rupiah'].str.replace(r'Rp', '', regex=True).astype('str')
data['price_rupiah'] = data['price_rupiah'].str.replace(r',', '', regex=True).astype('int')
```

```python
#create a new column price with dollars:

data["price_dollars"] = data["price_rupiah"]*0.000070
```

```python
# Replace the company names
data["company"] = data["company"].replace({"A": "Pizza Hut", "B": "Domino", "C": "Hunt Brothers Pizza", "D": "Little Caesars", "E" : "Papa John Pizza"})
```

```python
# Remove the old price column
data = data.drop(["price_rupiah"], axis = 1)
```
